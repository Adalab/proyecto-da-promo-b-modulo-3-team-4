```python
# Importamos librerías para el tratamiento de datos
import pandas as pd
import numpy as np
# Config pandas para ver todas las columnas 
pd.set_option("display.max_columns", None)


import mysql.connector

```

```python

# ABRIMOS CSV

df = pd.read_csv('data/HR RAW DATA_CLEAN.csv', index_col = 0)

print(df.shape)

df.head(2)

```

```python

# CREACION SCHEMA

# Nos conectamos a nuestro servidor local de SQL para poder hacer la creacion del Schema, las tablas y la inserción de datos

cnx = mysql.connector.connect(user = 'root',password = 'AlumnaAdalab',host = "127.0.0.1")
    
mycursor = cnx.cursor()

query_creacion_bbdd = "CREATE SCHEMA IF NOT EXISTS `Proyecto` DEFAULT CHARACTER SET utf8mb4;"
try:
    mycursor.execute(query_creacion_bbdd)
    print(mycursor)
    cnx.close()
# Si hubiera algun error, ejemplo contraseña incorrecta, indicamos que en lugar de saltar una excepcion se imprima el mensaje con el error en particular
except mysql.connector.Error as err:
    print(err)
    print(f"Error Code {err.errno}")
    print(f"SQLSTATE {err.sqlstate}")
    print(f"MESSAGE {err.msg}")
    cnx.close()

```

```python
# CREACION TABLA WORK


cnx = mysql.connector.connect(user = 'root',password = 'AlumnaAdalab',host = "127.0.0.1",database = 'Proyecto')

mycursor = cnx.cursor()

query_creacion_tabla_work = """
                            CREATE TABLE IF NOT EXISTS `work` (
                            `idWork` INT NOT NULL AUTO_INCREMENT,
                            `JobLevel` INT UNSIGNED NOT NULL,
                            `JobRole` VARCHAR(45) NOT NULL,
                            `Department` VARCHAR(45) NULL,
                            PRIMARY KEY (`idWork`))
                            ENGINE = InnoDB;
                            """
try:
    mycursor.execute(query_creacion_tabla_work)
    print(mycursor)
    
except mysql.connector.Error as err:
    print(err)
    print(f"Error Code {err.errno}")
    print(f"SQLSTATE {err.sqlstate}")
    print(f"MESSAGE {err.msg}")

cnx.close()
```

```python

# CREACION TABLA SATISFACTION SURVEYS

cnx = mysql.connector.connect(user = 'root',password = 'AlumnaAdalab',host = "127.0.0.1",database = 'Proyecto')

mycursor = cnx.cursor()

query_creacion_tabla_satisfaction_surveys = """
                                    CREATE TABLE IF NOT EXISTS `satisfaction_surveys` (
                                    `idSatisfaction` INT NOT NULL AUTO_INCREMENT,
                                    `JobInvolvement` INT UNSIGNED NULL,
                                    `PerformanceRating` INT UNSIGNED NULL,
                                    `EnvironmentSatisfaction` INT UNSIGNED NULL,
                                    `RelationshipSatisfaction` INT UNSIGNED NULL,
                                    `WorkLifeBalance` INT UNSIGNED NULL,
                                    `JobSatisfaction` INT UNSIGNED NULL,
                                    PRIMARY KEY (`idSatisfaction`))
                                    ENGINE = InnoDB;"""
try:
    mycursor.execute(query_creacion_tabla_satisfaction_surveys)
    print(mycursor)
    
except mysql.connector.Error as err:
    print(err)
    print(f"Error Code {err.errno}")
    print(f"SQLSTATE {err.sqlstate}")
    print(f"MESSAGE {err.msg}")

cnx.close()
        
```

```python

# CREACION TABLA EMPLOYEES

cnx = mysql.connector.connect(user = 'root',password = 'AlumnaAdalab',host = "127.0.0.1",database = 'Proyecto')

mycursor = cnx.cursor()

query_creacion_tabla_employees = """
                            CREATE TABLE IF NOT EXISTS `employees` (
                            `idemployee` INT NOT NULL AUTO_INCREMENT,
                            `EmployeeNumber` VARCHAR(45) NULL,
                            `Age` INT UNSIGNED NULL,
                            `Gender` VARCHAR(45) NULL,
                            `Education` INT UNSIGNED NULL,
                            `EducationField` VARCHAR(45) NULL,
                            `NumCompaniesWorked` INT UNSIGNED NULL,
                            `MonthlyIncome` INT UNSIGNED NULL,
                            `StockOptionLevel` INT NULL,
                            `YearsAtCompany` INT UNSIGNED NULL,
                            `YearsSinceLastPromotion` INT UNSIGNED NULL,
                            `YearsWithCurrManager` INT UNSIGNED NULL,
                            `TrainingTimesLastYear` INT UNSIGNED NULL,
                            `DistanceFromHome` INT UNSIGNED NULL,
                            `Attrition` VARCHAR(45) NOT NULL,
                            `idwork` INT NOT NULL,
                            `idSatisfaction` INT NOT NULL,
                            PRIMARY KEY (`idemployee`),
                            INDEX `fk_employees_work_idx` (`idwork` ASC) VISIBLE,
                            INDEX `fk_employees_Satisfaction_Surveys1_idx` (`idSatisfaction` ASC) VISIBLE,
                            CONSTRAINT `fk_employees_work`
                                FOREIGN KEY (`idwork`)
                                REFERENCES `proyecto`.`work` (`idwork`)
                                ON DELETE CASCADE
                                ON UPDATE CASCADE,
                            CONSTRAINT `fk_employees_Satisfaction_Surveys1`
                                FOREIGN KEY (`idSatisfaction`)
                                REFERENCES `proyecto`.`Satisfaction_Surveys` (`idSatisfaction`)
                                ON DELETE CASCADE
                                ON UPDATE CASCADE)
                            ENGINE = InnoDB;
                            """
try:
    mycursor.execute(query_creacion_tabla_employees)
    print(mycursor)
    
except mysql.connector.Error as err:
    print(err)
    print(f"Error Code {err.errno}")
    print(f"SQLSTATE {err.sqlstate}")
    print(f"MESSAGE {err.msg}")

cnx.close()
```

```python

# CREACION TABLA WORK_INFO

cnx = mysql.connector.connect(user = 'root',password = 'AlumnaAdalab',host = "127.0.0.1",database = 'Proyecto')

mycursor = cnx.cursor()

query_creacion_tabla_work_info =  """
                                CREATE TABLE IF NOT EXISTS `work_info` (
                                `idwork_info` INT NOT NULL AUTO_INCREMENT,
                                `RemoteWork` VARCHAR(45) NOT NULL,
                                `OverTime` VARCHAR(45) NULL,
                                `BusinessTravel` VARCHAR(45) NULL,
                                `idwork` INT NOT NULL,
                                PRIMARY KEY (`idwork_info`),
                                INDEX `fk_work_info_work1_idx` (`idwork` ASC) VISIBLE,
                                CONSTRAINT `fk_work_info_work1`
                                    FOREIGN KEY (`idwork`)
                                    REFERENCES `proyecto`.`work` (`idwork`)
                                    ON DELETE CASCADE
                                    ON UPDATE CASCADE)
                                ENGINE = InnoDB;
                                    """
try:
    mycursor.execute(query_creacion_tabla_work_info)
    print(mycursor)
    
except mysql.connector.Error as err:
    print(err)
    print(f"Error Code {err.errno}")
    print(f"SQLSTATE {err.sqlstate}")
    print(f"MESSAGE {err.msg}")

cnx.close()
```

```python
# INSERCION DATOS PRIMERO EN LAS TABLAS CUYA PRIMARY KEY ES FOREIGN KEY DE OTRA TABLA 

# INSERCION DATOS EN WORK, PARA LO CUAL PRIMERO EXTRAEMOS DEL DATAFRAME UNA LISTA QUE CONTIENE LA INFORMACION
# FILA A FILA EN TUPLAS, CADA TUPLA ES UNA FILA. 

valores_tabla_work = list(set(zip([int(x) for x in df['JobLevel'].values],df['JobRole'].values, df['Department'].values)))

print(valores_tabla_work)
```

```python
# UNA VEZ TENEMOS LOS VALORES LOS INSERTAMOS EN LA TABLA WORK


cnx = mysql.connector.connect(user = 'root',password = 'AlumnaAdalab',host = "127.0.0.1",database = 'Proyecto')

mycursor = cnx.cursor()
query_insercion_work = "INSERT INTO work (JobLevel, JobRole, Department) VALUES (%s, %s, %s)"

try:
    mycursor.executemany(query_insercion_work,valores_tabla_work)
    cnx.commit()
    print(mycursor.rowcount, "registros insertados")
    
except mysql.connector.Error as err:
    print(err)
    print(f"Error Code {err.errno}")
    print(f"SQLSTATE {err.sqlstate}")
    print(f"MESSAGE {err.msg}")

cnx.close()
```

```python
# HACEMOS EL MISMO PROCESO PARA LA TABLA SATISFACTION SURVEYS, PRIMERO EXTRAEMOS LOS DATOS EN LISTA DE TUPLAS. 
# USAMOS SET EN AMBOS CASOS PARA QUE NO HAYA CONJUNTO DE DATOS DUPLICADOS. 

valores_tabla_satisfaction = list(set(zip(
    [int(x) for x in df['JobInvolvement'].values],
    [float(x) for x in df['PerformanceRating'].values],
    [int(x) for x in df['EnvironmentSatisfaction'].values],
    [int(x) for x in df['RelationshipSatisfaction'].values],
    [float(x) for x in df['WorkLifeBalance'].values],
    [int(x) for x in df['JobSatisfaction'].values])))
    
print(valores_tabla_satisfaction)
```

```python
# INSERTAMOS LOS DATOS EN LA TABLA SATISFACTION SURVEYS

cnx = mysql.connector.connect(user = 'root',password = 'AlumnaAdalab',host = "127.0.0.1",database = 'Proyecto')

mycursor = cnx.cursor()
query_insercion_satisfaction = """INSERT INTO satisfaction_surveys (
                                    JobInvolvement, 
                                    PerformanceRating, 
                                    EnvironmentSatisfaction, 
                                    RelationshipSatisfaction, 
                                    WorkLifeBalance, 
                                    JobSatisfaction) 
                                    VALUES (%s, %s, %s, %s, %s, %s)"""

try:
    mycursor.executemany(query_insercion_satisfaction,valores_tabla_satisfaction)
    cnx.commit()
    print(mycursor.rowcount, "registros insertados")
    
except mysql.connector.Error as err:
    print(err)
    print(f"Error Code {err.errno}")
    print(f"SQLSTATE {err.sqlstate}")
    print(f"MESSAGE {err.msg}")

cnx.close()
```

```python
# PARA PODER RECOGER LOS DATOS NECESARIOS PARA EL DF EMPLOYEE HACEMOS UN FETCHALL PARA RECOGER LOS DATOS DE LAS TABLAS CREADAS
# Y PODER INCORPORARLO A NUESTRO DATAFRAME PARA HACER LA INSERCION EN EMPLOYEES

cnx = mysql.connector.connect(user = "root", password = 'AlumnaAdalab', host ="127.0.0.1", database = 'Proyecto')

mycursor = cnx.cursor()

query = ("SELECT * FROM work")

mycursor.execute(query)

datos_work = mycursor.fetchall()
    
cnx.close()
    
dataframe_work = pd.DataFrame(datos_work, columns = ['idWork', 'JobLevel', 'JobRole', 'Department'])
     
df_employees_work = pd.merge(df, dataframe_work, on = ['JobLevel', 'JobRole', 'Department'])

print("filas y columnas", df_employees_work.shape)

df_employees_work
    
```

```python
# HACEMOS LO MISMO PARA RECOGER LOS DATOS DE SATISFACTION SURVEYS

cnx = mysql.connector.connect(user = "root", password = 'AlumnaAdalab', host ="127.0.0.1", database = 'Proyecto')

mycursor = cnx.cursor()

query = ("SELECT * FROM satisfaction_surveys")

mycursor.execute(query)

datos_satisfaction = mycursor.fetchall()
    
cnx.close()
    
dataframe_satisfaction = pd.DataFrame(datos_satisfaction, columns = ['idSatisfaction', 'JobInvolvement', 'PerformanceRating', 'EnvironmentSatisfaction', 'RelationshipSatisfaction', 'WorkLifeBalance', 'JobSatisfaction'])
     
df_employees_completo = pd.merge(df_employees_work, dataframe_satisfaction, on = ['JobInvolvement', 'PerformanceRating', 'EnvironmentSatisfaction', 'RelationshipSatisfaction', 'WorkLifeBalance', 'JobSatisfaction'])

print("filas y columnas", df_employees_completo.shape)

df_employees_completo
```

```python
# RECOGEMOS LOS VALORES QUE NOS INTERESAN DEL DATAFRAME COMPLETO PARA INSERTARLOS EN LA TABLA EMPLOYEES

valores_tabla_employee = list(zip(
    df_employees_completo['EmployeeNumber'].values,
    [int(x) for x in df_employees_completo['Age'].values],
    df_employees_completo['Gender'].values,   
    [int(x) for x in df_employees_completo['Education'].values], 
    df_employees_completo['EducationField'].values, 
    [int(x) for x in df_employees_completo['NumCompaniesWorked'].values],
    [float(x) for x in df_employees_completo['MonthlyIncome'].values],
    [int(x) for x in df_employees_completo['StockOptionLevel'].values],
    [int(x) for x in df_employees_completo['YearsAtCompany'].values],
    [int(x) for x in df_employees_completo['YearsSinceLastPromotion'].values], 
    [int(x) for x in df_employees_completo['YearsWithCurrManager'].values],
    [int(x) for x in df_employees_completo['TrainingTimesLastYear'].values], 
    [int(x) for x in df_employees_completo['DistanceFromHome'].values],
    df_employees_completo['Attrition'].values, 
    [int(x) for x in df_employees_completo['idWork'].values],
    [int(x) for x in df_employees_completo['idSatisfaction']]
))

print(valores_tabla_employee)
```

```python
# INSERTAMOS LOS DATOS EN LA TABLA EMPLOYEE
cnx = mysql.connector.connect(user = 'root',password = 'AlumnaAdalab',host = "127.0.0.1",database = 'Proyecto')

mycursor = cnx.cursor()
query_insercion_employees = """INSERT INTO employees (
                                    EmployeeNumber, 
                                    Age, 
                                    Gender, 
                                    Education, 
                                    EducationField, 
                                    NumCompaniesWorked, 
                                    MonthlyIncome, 
                                    StockOptionLevel,
                                    YearsAtCompany, 
                                    YearsSinceLastPromotion,
                                    YearsWithCurrManager,
                                    TrainingTimesLastYear,                               
                                    DistanceFromHome, 
                                    Attrition,
                                    idWork,
                                    idSatisfaction) 
                                    VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)"""

try:
    mycursor.executemany(query_insercion_employees,valores_tabla_employee)
    cnx.commit()
    print(mycursor.rowcount, "registros insertados")
    
except mysql.connector.Error as err:
    print(err)
    print(f"Error Code {err.errno}")
    print(f"SQLSTATE {err.sqlstate}")
    print(f"MESSAGE {err.msg}")

cnx.close()
```
